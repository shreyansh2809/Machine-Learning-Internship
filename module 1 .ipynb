```python
def vwap(mon,yy):
    num=(df[(df['Month']==mon) & (df['Year']==yy)]['Close Price'] * df[(df['Month']==mon) & (df['Year']==yy)]['Total Traded Quantity']).sum()
    den=df[(df['Month']==mon) & (df['Year']==yy)]['Total Traded Quantity'].sum()
    return num/den
```

```python
new_df['VWAP'] = new_df['group'].apply(lambda x:vwap(x[0],x[1]))
```

<h1>1.5 Write a function to calculate Average Price Over N days and Profit/Loss % over N days</h1>

```python
def lastN_Days(n):
    out=0
    l=list(df.index)
    l.sort(reverse = True)
    for i in l:
        if (df['Date'][495]-df['Date'][i]).days == n :
            out = i
            break
        elif (df['Date'][495]-df['Date'][i]).days > n :
            out = i+1
            break
    return out
def avg_price_lastN_Days(n):
    id=lastN_Days(n)
    last = 495 - id
    return df['Average Price'].tail(last).sum()/n
```

```python
print("Average Price Over Last N Days : ")
print("1 Week : ",avg_price_lastN_Days(7))
print("2 Week : ",avg_price_lastN_Days(14))
print("1 Month : ",avg_price_lastN_Days(30))
print("3 Months : ",avg_price_lastN_Days(90))
print("6 Months : ",avg_price_lastN_Days(180))
print("1 Year : ",avg_price_lastN_Days(365))
```

```python
def pro_loss(n):
    id=lastN_Days(n)
    diff = df['Close Price'][495] - df['Close Price'][id]
    prolo = 0
    if diff < 0:
        #loss
        prolo = (((-1)*diff) / df['Close Price'][id] )*100
        return 1,prolo
    else:
        prolo = ((diff) / df['Close Price'][id] )*100
        return 0,prolo
l=['Profit% :','Loss% :']
        
```

```python
print("Profit / Loss % over last : ")
print("1 Week : ",l[pro_loss(7)[0]],pro_loss(7)[1])
print("2 Week : ",l[pro_loss(14)[0]],pro_loss(14)[1])
print("1 Month : ",l[pro_loss(30)[0]],pro_loss(30)[1])
print("3 Month : ",l[pro_loss(90)[0]],pro_loss(90)[1])
print("6 Month : ",l[pro_loss(180)[0]],pro_loss(180)[1])
print("1 Year : ",l[pro_loss(365)[0]],pro_loss(365)[1])
```

<h1>1.7 Add another Column 'Trend'</h1>

```python
def day_per_change_comment(x):
    if x>=-0.5*0.01 and x<=0.5*0.01:
        return 'Slight or No change'
    elif x>=0.5*0.01 and x<=1.0*0.01:
        return 'Slight positive'
    elif x>=-1.0*0.01 and x<=-0.5*0.01:
        return 'Slight negative'
    elif x>=1*0.01 and x<=3*0.01:
        return 'Positive'
    elif x>=-3*0.01 and x<=-1*0.01:
        return 'Negative'
    elif x>=3*0.01 and x<=7*0.01:
        return 'Among top gainers'
    elif x>=-7*0.01 and x<=-3*0.01:
        return 'Among top losers'
    elif x>7*0.01:
        return 'Bull run'
    elif x<-7*0.01:
        return 'Bear drop'
```

```python
df['Trend'] = df['Day_Perc_Change'].apply(day_per_change_comment)
```
