<h2>2.2 Display a stem plot of daily change in the stock price in percentage </h2>

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.stem(df.index,df['Day_Perc_Change'],use_line_collection=True,bottom=0)
ax.set_xlabel('Date')
ax.set_ylabel('Day_Perc_Change')
ax.set_title("Stem Plot")
```

<h2>2.3 Plot the Daily Volumes as well as compare the percentage stem plot to it.</h2> 

```python
#plt.plot(df.index,df['No. of Trades'], figsize=(10,8))

fig = plt.figure(figsize=(10,8))
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(df.index,df['No. of Trades'])
ax1.set_xlabel('Date')
ax1.set_ylabel('Volume')
```

```python
sns.set(style="darkgrid")


scaledvolume =  df["No. of Trades"] - df["No. of Trades"].min()
scaledvolume = scaledvolume/scaledvolume.max() * df.Day_Perc_Change.max()

fig, ax2 = plt.subplots(figsize=(12, 6))

ax2.stem(df.index, df.Day_Perc_Change , 'b', markerfmt='bo', label='Percente Change',use_line_collection=True)
ax2.plot(df.index, scaledvolume, 'k', label='Volume')

ax2.set_xlabel('Date')
plt.legend(loc=2)

plt.tight_layout()
plt.xticks(plt.xticks()[0], df.index.date, rotation=45)
plt.show()
```

<h2>2.9 Trade Calls Using simple moving averages</h2>

```python
avg21 = df['Average Price'].rolling(21).mean()
avg34 = df['Average Price'].rolling(34).mean()
```

```python
plt.figure(figsize=(20,10))
plt.grid(True)
avg21.plot(color = 'red',alpha = 0.5,label='Moving Average 21')
avg34.plot(color = 'blue',alpha = 0.5,label='Moving Average 34')
df['Average Price'].plot(color = 'green',label = 'Average Price',linewidth = 1.5)
plt.legend()
plt.title("Average Price")
```

<h2>2.10 Bollinger Band</h2>

```python
df['14 Day MA'] = df['Close Price'].rolling(14).mean()
df['14 Day STD'] = df['Close Price'].rolling(14).std()
df['Upper Band'] = df['14 Day MA'] + (df['14 Day STD'] * 2)
df['Lower Band'] = df['14 Day MA'] - (df['14 Day STD'] * 2)
```

```python
plt.figure(figsize=(20,10))
plt.grid(True)
df['14 Day MA'].plot(color = 'red',alpha = 0.5,label='14 Day Mean Average')
df['Upper Band'].plot(color = 'blue',alpha = 0.5,label='Upper Band')
df['Lower Band'].plot(color = 'green',alpha = 0.5,label = 'Lower Band')
df['Average Price'].plot(color = 'violet',alpha = 0.5,label = 'Average Daily Stock Price')
plt.legend()
plt.title("Bollinger Band")
```
